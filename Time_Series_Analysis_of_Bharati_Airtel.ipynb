```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
import numpy as np

# Step 2: Load Dataset
data = pd.read_csv('BhartiAirtel.csv')  # Update the path if needed
# Original line causing the error:
# data['Date'] = pd.to_datetime(data['Date'], format='%d-%b-%y')

# Corrected line:
data['Date'] = pd.to_datetime(data['Date'], format='%d-%B-%Y')
# Changed %b to %B for full month name
# Changed %y to %Y for full year
data.set_index('Date', inplace=True)
df = data['Close Price']

# Step 3: Plot the Close Price
plt.figure(figsize=(12,6))
plt.plot(df, label='Close Price')
plt.title('Close Price over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Step 4: Check for Stationarity
# Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] < 0.05:
        print("The series is stationary.")
    else:
        print("The series is not stationary.")

adf_test(df)

# Step 5: Differencing if Needed
# Differencing to make the data stationary if p-value > 0.05
df_diff = df.diff().dropna()
adf_test(df_diff)  # Re-test for stationarity

# Step 6: Model Selection and Fitting
# Method 1: ARIMA Model
# Auto-regressive Integrated Moving Average model

# Choosing ARIMA orders (p, d, q)
# Here we assume differencing (d=1) based on previous stationarity check
p, d, q = 1, 1, 1  # Simple model for demonstration

arima_model = ARIMA(df, order=(p, d, q))
arima_fit = arima_model.fit()
print(arima_fit.summary())

# Step 7: Forecasting with ARIMA
forecast_arima = arima_fit.forecast(steps=10)
plt.figure(figsize=(12,6))
plt.plot(df, label='Close Price')
plt.plot(forecast_arima, label='ARIMA Forecast', color='red')
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Method 2: Exponential Smoothing (Holt-Winters)
# This is good for data with trend and/or seasonality
ets_model = ExponentialSmoothing(df, trend='add', seasonal=None, seasonal_periods=12)
ets_fit = ets_model.fit()
forecast_ets = ets_fit.forecast(steps=10)

# Plotting ETS forecast
plt.figure(figsize=(12,6))
plt.plot(df, label='Close Price')
plt.plot(forecast_ets, label='ETS Forecast', color='green')
plt.title('ETS Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

# Step 8: Evaluate the Models
# Calculating RMSE for both ARIMA and ETS
actual = df[-10:]
arima_rmse = np.sqrt(mean_squared_error(actual, forecast_arima))
ets_rmse = np.sqrt(mean_squared_error(actual, forecast_ets))
print(f"ARIMA RMSE: {arima_rmse}")
print(f"ETS RMSE: {ets_rmse}")
```
